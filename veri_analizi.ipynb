yılları tespit et, sütunları temizle, sayısallık sağlamak

```python
# Product ve GeoAreaName sütunlarını string olarak normalize et
df['Product'] = df['Product'].astype(str).str.strip()
df['GeoAreaName'] = df['GeoAreaName'].astype(str).str.strip()
if 'Flow' in df.columns:
    df['Flow'] = df['Flow'].astype(str).str.strip()

# Yılları otomatik tespit et (sadece dört haneli yıl isimleri)
year_cols = [c for c in df.columns if isinstance(c, str) and c.isdigit() and 1900 <= int(c) <= 2100]
year_cols = sorted(year_cols, key=lambda x: int(x))
print("Year columns:", year_cols)

# Yılları sayısala çevir (NaN oluşturabilecek metinleri float'a)
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
```

Product == "Renewable" (tam eşleşme) için yıllara göre toplam çizgisi

```python
product_name = "Renewables"   # tam eşleşecek şekilde yaz

df_prod = df[df['Product'] == product_name].copy()

if df_prod.empty:
    print(f"'{product_name}' adına tam eşleşen satır bulunamadı. Mevcut product örnekleri:\n", df['Product'].unique()[:50])
else:
    renewable_yearly = df_prod[year_cols].sum(axis=0, skipna=True)
    plt.figure(figsize=(12,6))
    plt.plot(list(map(int, renewable_yearly.index)), renewable_yearly.values, marker='o')
    plt.title(f"Toplam {product_name} Enerji (Yıllara Göre)")
    plt.xlabel("Yıl")
    plt.ylabel("Değer (toplam)")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

Ülkelere göre toplam Product miktarı (bar chart, tam eşleşme)

```python
product_name = "Renewables"   # istediğin product adı

df_prod = df[df['Product'] == product_name].copy()
df_prod['Total'] = df_prod[year_cols].sum(axis=1, skipna=True)

country_totals = df_prod.groupby("GeoAreaName")['Total'].sum().sort_values(ascending=False)

plt.figure(figsize=(14,14))
country_totals.head(20).plot(kind='bar')
plt.title(f"Ülkelere Göre Toplam '{product_name}' (En üst 20)")
plt.xlabel("Ülke")
plt.ylabel("Toplam Değer")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Bir ülke için tüm Product türlerinin zaman içindeki değişimi (tam eşleşme ile product seçimi)

```python
country = "China"   
df_country = df[df['GeoAreaName'] == country].copy()

if df_country.empty:
    print(f"Ülke '{country}' bulunamadı. Mevcut örnek ülke isimleri:\n", df['GeoAreaName'].unique()[:50])
else:
    plt.figure(figsize=(12,12))
    for product in df_country['Product'].unique():
        sub = df_country[df_country['Product'] == product]
        # Eğer aynı product için birden fazla satır varsa toplam al
        series = sub[year_cols].sum(axis=0, skipna=True)
        plt.plot(list(map(int, series.index)), series.values, label=product)
    plt.title(f"{country} — Ürünlere Göre Zaman Serileri")
    plt.xlabel("Yıl")
    plt.ylabel("Değer")
    plt.legend(loc='best', fontsize='small', ncol=2)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### Product kategorilerinin toplam payı (pie chart)

```python
df['Total'] = df[year_cols].sum(axis=1, skipna=True)

filtered = df[
    (df['Product'] != "Total") & 
    (df['Product'] != "Renewables")
]

product_totals = (
    filtered.groupby("Product")["Total"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,35))
product_totals.head(10).plot(kind="pie", autopct="%1.1f%%")
plt.title("Ürün Kategorilerinin Toplam Payı (Total ve Renewables Hariç)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
